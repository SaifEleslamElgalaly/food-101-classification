```python
import os
import shutil

# المسارات
original_images_path = "D:/Education/CIC/semster 6/Pattern Recogonition/Project/food-101/images"
new_dataset_path = "D:/Education/CIC/semster 6/Pattern Recogonition/Project/food-101-reduced"

# إنشاء مجلد جديد لحفظ الـ dataset المصغرة
os.makedirs(new_dataset_path, exist_ok=True)

# قائمة الـ classes المختارة
selected_classes = [
    "pizza", "cheeseburger", "french_fries", "tacos", "ramen",
    "caesar_salad", "sushi", "donuts", "cheesecake", "fried_chicken",
    "bread_pudding", "apple_pie", "steak", "grilled_salmon", "milkshake"
]

# نقل الصور المطلوبة فقط
for class_name in selected_classes:
    src = os.path.join(original_images_path, class_name)
    dst = os.path.join(new_dataset_path, class_name)
    if os.path.exists(src):  # تأكد أن المجلد موجود
        shutil.copytree(src, dst)
        print(f"✅ تم نقل {class_name}")
    else:
        print(f"⚠️ المجلد {class_name} غير موجود في البيانات الأصلية!")

print("🎯 تم إنشاء الـ dataset المصغرة بنجاح!")
```

```python
import os
import json

# المسارات
meta_path = "D:/Education/CIC/semster 6/Pattern Recogonition/Project/food-101/meta"
new_meta_path = "D:/Education/CIC/semster 6/Pattern Recogonition/Project/food-101-reduced/meta"
os.makedirs(new_meta_path, exist_ok=True)

# قائمة الـ classes المختارة
selected_classes = [
    "pizza", "waffles","samosa","ice_cream", "french_fries", "tacos", "ramen",
    "caesar_salad", "sushi", "donuts", "cheesecake",
    "bread_pudding", "apple_pie", "steak", "grilled_salmon"
]

### تحديث ملف classes.txt
classes_file = os.path.join(meta_path, "classes.txt")
new_classes_file = os.path.join(new_meta_path, "classes.txt")

with open(classes_file, "r") as f:
    all_classes = f.read().splitlines()

filtered_classes = [cls for cls in all_classes if cls in selected_classes]

with open(new_classes_file, "w") as f:
    f.write("\n".join(filtered_classes))

print("✅ تم تحديث ملف classes.txt")

### تحديث train.txt و test.txt
for file_name in ["train.txt", "test.txt"]:
    input_file = os.path.join(meta_path, file_name)
    output_file = os.path.join(new_meta_path, file_name)

    with open(input_file, "r") as f:
        lines = f.read().splitlines()

    filtered_lines = [line for line in lines if line.split("/")[0] in selected_classes]

    with open(output_file, "w") as f:
        f.write("\n".join(filtered_lines))

    print(f"✅ تم تحديث ملف {file_name}")

### تحديث train.json و test.json
for file_name in ["train.json", "test.json"]:
    input_file = os.path.join(meta_path, file_name)
    output_file = os.path.join(new_meta_path, file_name)

    with open(input_file, "r") as f:
        data = json.load(f)

    filtered_data = {cls: images for cls, images in data.items() if cls in selected_classes}

    with open(output_file, "w") as f:
        json.dump(filtered_data, f, indent=4)

    print(f"✅ تم تحديث ملف {file_name}")

print("🎯 تم تحديث جميع ملفات meta بنجاح!")
```

```python
import json

for split in ["train", "test"]:
    # تحميل البيانات من JSON
    with open(f"meta/{split}.json", "r") as f:
        json_data = json.load(f)

    # تحميل البيانات من TXT
    with open(f"meta/{split}.txt", "r") as f:
        txt_data = [line.strip() for line in f.readlines()]

    # حساب عدد الصور لكل فئة
    json_counts = {}
    for img_path in json_data.keys():
        category = img_path.split("/")[0]
        json_counts[category] = json_counts.get(category, 0) + 1

    txt_counts = {}
    for line in txt_data:
        category = line.split("/")[0]
        txt_counts[category] = txt_counts.get(category, 0) + 1

    # التحقق من المطابقة
    print(f"\n📊 مقارنة عدد الصور في {split}.txt و {split}.json:")
    for category in sorted(set(json_counts.keys()).union(txt_counts.keys())):
        json_count = json_counts.get(category, 0)
        txt_count = txt_counts.get(category, 0)
        status = "✅ متطابق" if json_count == txt_count else "❌ غير متطابق"
        print(f"{category}: JSON={json_count} | TXT={txt_count} → {status}")
```

```python
import json
from collections import Counter

# التحقق من البيانات في train.txt و train.json
for split in ["train", "test"]:
    # قراءة train.txt و test.txt
    txt_counts = Counter()
    with open(f"meta/{split}.txt", "r") as f:
        for line in f:
            class_name = line.strip().split("/")[0]
            txt_counts[class_name] += 1

    # قراءة train.json و test.json
    with open(f"meta/{split}.json", "r") as f:
        json_data = json.load(f)

    json_counts = {class_name: len(images) for class_name, images in json_data.items()}

    # مقارنة الأعداد
    print(f"\n📊 مقارنة عدد الصور في {split}.txt و {split}.json:")
    for class_name in sorted(txt_counts.keys()):
        txt_count = txt_counts[class_name]
        json_count = json_counts.get(class_name, 0)
        status = "✅" if txt_count == json_count else "❌"
        print(f"{status} {class_name}: JSON={json_count} | TXT={txt_count}")
```

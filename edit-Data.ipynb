```python
import os
import json

# Ø§Ù„Ù…Ø³Ø§Ø±Ø§Øª
meta_path = "D:/Education/CIC/semster 6/Pattern Recogonition/Project/food-101/meta"
new_meta_path = "D:/Education/CIC/semster 6/Pattern Recogonition/Project/food-101-reduced/meta"
os.makedirs(new_meta_path, exist_ok=True)

# Ù‚Ø§Ø¦Ù…Ø© Ø§Ù„Ù€ classes Ø§Ù„Ù…Ø®ØªØ§Ø±Ø©
selected_classes = [
    "pizza", "waffles","samosa","ice_cream", "french_fries", "tacos", "ramen",
    "caesar_salad", "sushi", "donuts", "cheesecake",
    "bread_pudding", "apple_pie", "steak", "grilled_salmon"
]

### ØªØ­Ø¯ÙŠØ« Ù…Ù„Ù classes.txt
classes_file = os.path.join(meta_path, "classes.txt")
new_classes_file = os.path.join(new_meta_path, "classes.txt")

with open(classes_file, "r") as f:
    all_classes = f.read().splitlines()

filtered_classes = [cls for cls in all_classes if cls in selected_classes]

with open(new_classes_file, "w") as f:
    f.write("\n".join(filtered_classes))

print("âœ… ØªÙ… ØªØ­Ø¯ÙŠØ« Ù…Ù„Ù classes.txt")

### ØªØ­Ø¯ÙŠØ« train.txt Ùˆ test.txt
for file_name in ["train.txt", "test.txt"]:
    input_file = os.path.join(meta_path, file_name)
    output_file = os.path.join(new_meta_path, file_name)

    with open(input_file, "r") as f:
        lines = f.read().splitlines()

    filtered_lines = [line for line in lines if line.split("/")[0] in selected_classes]

    with open(output_file, "w") as f:
        f.write("\n".join(filtered_lines))

    print(f"âœ… ØªÙ… ØªØ­Ø¯ÙŠØ« Ù…Ù„Ù {file_name}")

### ØªØ­Ø¯ÙŠØ« train.json Ùˆ test.json
for file_name in ["train.json", "test.json"]:
    input_file = os.path.join(meta_path, file_name)
    output_file = os.path.join(new_meta_path, file_name)

    with open(input_file, "r") as f:
        data = json.load(f)

    filtered_data = {cls: images for cls, images in data.items() if cls in selected_classes}

    with open(output_file, "w") as f:
        json.dump(filtered_data, f, indent=4)

    print(f"âœ… ØªÙ… ØªØ­Ø¯ÙŠØ« Ù…Ù„Ù {file_name}")

print("ğŸ¯ ØªÙ… ØªØ­Ø¯ÙŠØ« Ø¬Ù…ÙŠØ¹ Ù…Ù„ÙØ§Øª meta Ø¨Ù†Ø¬Ø§Ø­!")
```

```python
import json
from collections import Counter

# Ø§Ù„ØªØ­Ù‚Ù‚ Ù…Ù† Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª ÙÙŠ train.txt Ùˆ train.json
for split in ["train", "test"]:
    # Ù‚Ø±Ø§Ø¡Ø© train.txt Ùˆ test.txt
    txt_counts = Counter()
    with open(f"meta/{split}.txt", "r") as f:
        for line in f:
            class_name = line.strip().split("/")[0]
            txt_counts[class_name] += 1

    # Ù‚Ø±Ø§Ø¡Ø© train.json Ùˆ test.json
    with open(f"meta/{split}.json", "r") as f:
        json_data = json.load(f)

    json_counts = {class_name: len(images) for class_name, images in json_data.items()}

    # Ù…Ù‚Ø§Ø±Ù†Ø© Ø§Ù„Ø£Ø¹Ø¯Ø§Ø¯
    print(f"\nğŸ“Š Ù…Ù‚Ø§Ø±Ù†Ø© Ø¹Ø¯Ø¯ Ø§Ù„ØµÙˆØ± ÙÙŠ {split}.txt Ùˆ {split}.json:")
    for class_name in sorted(txt_counts.keys()):
        txt_count = txt_counts[class_name]
        json_count = json_counts.get(class_name, 0)
        status = "âœ…" if txt_count == json_count else "âŒ"
        print(f"{status} {class_name}: JSON={json_count} | TXT={txt_count}")
```

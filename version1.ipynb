```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python

# Ù†Ø­Ø¯Ø¯ Ø§Ù„Ù…Ø³Ø§Ø±Ø§Øª
train_dir = os.path.join(data_path, 'train')
valid_dir = os.path.join(data_path, 'valid')
test_dir  = os.path.join(data_path, 'test')

# ImageDataGenerator Ø¨Ø¯ÙˆÙ† Ø£ÙŠ Augmentation
datagen = ImageDataGenerator(rescale=1./255)

# ØªØ­Ù…ÙŠÙ„ Ø§Ù„ØµÙˆØ± Ù…Ù† Ø§Ù„Ù…Ø¬Ù„Ø¯Ø§Øª Ø¨Ø§Ù„Ø­Ø¬Ù… Ø§Ù„Ø¬Ø¯ÙŠØ¯
train_data = datagen.flow_from_directory(train_dir, target_size=(224, 224), batch_size=32, class_mode='categorical')
valid_data = datagen.flow_from_directory(valid_dir, target_size=(224, 224), batch_size=32, class_mode='categorical')
test_data  = datagen.flow_from_directory(test_dir,  target_size=(224, 224), batch_size=32, class_mode='categorical')

```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Load the MobileNetV2 model without the top layers
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')


# Freeze the base model so we only train the new layers
base_model.trainable = False

# Add our own layers on top
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
predictions = Dense(15, activation='softmax')(x)

# Final model
model = Model(inputs=base_model.input, outputs=predictions)

# Compile
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Summary
model.summary()
```

```python
history = model.fit(
    train_data,
    validation_data=valid_data,
    epochs=10
)
```
